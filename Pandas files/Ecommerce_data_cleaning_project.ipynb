# *Alternate way of dealing null values by automating the forward and backward filling process*

```python
# making a copy of the main dataframe
df3=df_ecom.copy()
df3
```

```python
#sorting the values based on product
df3=df3.sort_values('Product',ignore_index=True)
df3
```

```python
# automating the filling process uaing for loop and if statement.this would fill forward value or backward value based on the conditions given, thus filling up all the null values in one go.
for i in df3.index:
    if type(df3.loc[i,'Category'])==str:
        continue
    elif df3.loc[i,'Product']==df3.loc[i+1,'Product']:
        df3.loc[i,'Category']=df3.loc[i+1,'Category']
    elif  (i!=0) & (df3.loc[i,'Product']==df3.loc[i-1,'Product']):
        df3.loc[i,'Category']=df3.loc[i-1,'Category']
     
df3       
```

```python
# reverting back to the original sort order of the dataframe
df3=df3.sort_values('ID',ignore_index=True)
df3
```

```python
#crosschecking the null values after applying the above changes
df3.isna().sum()
```
